```python
# Sample Data for Main Kitchen and Secondary Kitchen
main_kitchen = [
    ("Tomatoes", 5, "Vegetable"),
    ("Potatoes", 10, "Vegetable"),
    ("Chicken", 2, "Meat"),
    ("Rice", 1, "Grain"),
    ("Milk", 2, "Dairy")
]

secondary_kitchen = [
    ("Eggs", 12, "Dairy"),
    ("Potatoes", 5, "Vegetable"),  # Common item
    ("Chicken", 1, "Meat"),        # Common item
    ("Bread", 3, "Grain"),
    ("Cheese", 1, "Dairy")
]

# Functions for various operations
def add_item():
    name = input("Enter Item Name: ")
    quantity = int(input("Enter Quantity: "))
    category = input("Enter Category (e.g., Vegetable, Meat, Dairy): ")

    new_item = (name, quantity, category)
    main_kitchen.append(new_item)
    print("\nItem Added Successfully!")

def display_items():
    print("\nItems in Main Kitchen:")
    for item in main_kitchen:
        print(f"Item: {item[0]}, Quantity: {item[1]}, Category: {item[2]}")

def remove_duplicates():
    global main_kitchen
    main_kitchen = list(set(main_kitchen))
    print("\nDuplicates Removed!")

def find_common_items():
    common_items = set(main_kitchen).intersection(set(secondary_kitchen))
    print("\nCommon Items between Main and Secondary Kitchen:")
    for item in common_items:
        print(f"Item: {item[0]}, Quantity: {item[1]}, Category: {item[2]}")

def find_unique_to_main_kitchen():
    unique_to_main = set(main_kitchen).difference(set(secondary_kitchen))
    print("\nItems Unique to Main Kitchen:")
    for item in unique_to_main:
        print(f"Item: {item[0]}, Quantity: {item[1]}, Category: {item[2]}")

# Main function to provide a menu for the user
def main():
    while True:
        print("\n--- Kitchen Inventory Management ---")
        print("1. Add Item")
        print("2. Display Items")
        print("3. Remove Duplicates")
        print("4. Find Common Items (Main & Secondary)")
        print("5. Find Unique Items to Main Kitchen")
        print("6. Exit")

        choice = input("Enter your choice (1-6): ")

        if choice == '1':
            add_item()
        elif choice == '2':
            display_items()
        elif choice == '3':
            remove_duplicates()
        elif choice == '4':
            find_common_items()
        elif choice == '5':
            find_unique_to_main_kitchen()
        elif choice == '6':
            print("Exiting the program.")
            break
        else:
            print("Invalid choice! Please enter a number between 1 and 6.")

# Run the main function
if __name__ == "__main__":
    main()
```

```python
import gradio as gr

# Sample data for Main Kitchen and Secondary Kitchen
main_kitchen = [
    ("Tomatoes", 5, "Vegetable"),
    ("Potatoes", 10, "Vegetable"),
    ("Chicken", 2, "Meat"),
    ("Rice", 1, "Grain"),
    ("Milk", 2, "Dairy")
]

secondary_kitchen = [
    ("Eggs", 12, "Dairy"),
    ("Potatoes", 5, "Vegetable"),  # Common item
    ("Chicken", 1, "Meat"),        # Common item
    ("Bread", 3, "Grain"),
    ("Cheese", 1, "Dairy")
]

# Define the available categories
categories = ["Vegetable", "Meat", "Grain", "Dairy", "Fruit"]
kitchen_choices = ["Main Kitchen", "Secondary Kitchen"]

# Function to add a new item to the selected kitchen inventory
def add_item(name, quantity, category, kitchen_choice):
    new_item = (name, quantity, category)

    if kitchen_choice == "Main Kitchen":
        main_kitchen.append(new_item)
        return f"Item '{name}' added to Main Kitchen successfully!"
    elif kitchen_choice == "Secondary Kitchen":
        secondary_kitchen.append(new_item)
        return f"Item '{name}' added to Secondary Kitchen successfully!"

# Function to display all items in the selected kitchen inventory
def display_items(kitchen_choice):
    if kitchen_choice == "Main Kitchen":
        kitchen = main_kitchen
    else:
        kitchen = secondary_kitchen

    if kitchen:
        items = "\n".join([f"Item: {item[0]}, Quantity: {item[1]}, Category: {item[2]}" for item in kitchen])
        return f"Items in {kitchen_choice}:\n{items}"
    else:
        return f"{kitchen_choice} is empty!"

# Function to remove duplicates from the selected kitchen inventory
def remove_duplicates(kitchen_choice):
    if kitchen_choice == "Main Kitchen":
        seen = {}
        global main_kitchen
        for item in main_kitchen:
            if item[0] not in seen:
                seen[item[0]] = item
        main_kitchen = list(seen.values())
        return "Duplicates removed successfully from Main Kitchen!"
    elif kitchen_choice == "Secondary Kitchen":
        seen = {}
        global secondary_kitchen
        for item in secondary_kitchen:
            if item[0] not in seen:
                seen[item[0]] = item
        secondary_kitchen = list(seen.values())
        return "Duplicates removed successfully from Secondary Kitchen!"

# Function to find common items between main and secondary kitchen inventories
def find_common_items():
    common_items = set(main_kitchen).intersection(set(secondary_kitchen))
    if common_items:
        items = "\n".join([f"Item: {item[0]}, Quantity: {item[1]}, Category: {item[2]}" for item in common_items])
        return f"Common Items between Main and Secondary Kitchen:\n{items}"
    else:
        return "No common items found."

# Function to find unique items based on the selected kitchen
def find_unique_items(kitchen_choice):
    if kitchen_choice == "Main Kitchen":
        unique_items = set(main_kitchen).difference(set(secondary_kitchen))
        if unique_items:
            items = "\n".join([f"Item: {item[0]}, Quantity: {item[1]}, Category: {item[2]}" for item in unique_items])
            return f"Items Unique to Main Kitchen:\n{items}"
        else:
            return "No unique items in Main Kitchen."
    else:  # For Secondary Kitchen
        unique_items = set(secondary_kitchen).difference(set(main_kitchen))
        if unique_items:
            items = "\n".join([f"Item: {item[0]}, Quantity: {item[1]}, Category: {item[2]}" for item in unique_items])
            return f"Items Unique to Secondary Kitchen:\n{items}"
        else:
            return "No unique items in Secondary Kitchen."

# Gradio Interface
def kitchen_interface(name, quantity, category, kitchen_choice):
    add_message = add_item(name, quantity, category, kitchen_choice)
    display_message = display_items(kitchen_choice)  # Show items from the selected kitchen after adding
    return add_message, display_message

def display_kitchen_items(kitchen_choice):
    return display_items(kitchen_choice)

with gr.Blocks() as app:
    gr.Markdown("## Kitchen Inventory Management")

    with gr.Row():
        with gr.Column():
            gr.Markdown("### Add New Item")
            name_input = gr.Textbox(label="Item Name")
            quantity_input = gr.Number(label="Quantity", value=1)
            category_input = gr.Dropdown(choices=categories, label="Category")  # Dropdown for categories
            kitchen_choice_input = gr.Dropdown(choices=kitchen_choices, label="Select Kitchen", value="Main Kitchen")  # Dropdown for kitchen choice
            add_message = gr.Textbox(label="Add Item Status", interactive=False)
            display_message = gr.Textbox(label="Current Inventory", interactive=False)  # Define this before using

            add_button = gr.Button("Add Item")
            add_button.click(kitchen_interface, inputs=[name_input, quantity_input, category_input, kitchen_choice_input], outputs=[add_message, display_message])

    with gr.Row():
        gr.Markdown("### Show Items in Kitchen")
        display_button = gr.Button("Show Current Items")
        display_button.click(display_kitchen_items, inputs=kitchen_choice_input, outputs=display_message)

    with gr.Row():
        remove_kitchen_choice_input = gr.Dropdown(choices=kitchen_choices, label="Select Kitchen for Duplicates", value="Main Kitchen")
        remove_button = gr.Button("Remove Duplicates")
        remove_message = gr.Textbox(label="Remove Duplicates Status", interactive=False)
        remove_button.click(remove_duplicates, inputs=remove_kitchen_choice_input, outputs=remove_message)

    with gr.Row():
        common_button = gr.Button("Find Common Items")
        common_message = gr.Textbox(label="Common Items Status", interactive=False)
        common_button.click(find_common_items, outputs=common_message)

    with gr.Row():
        unique_kitchen_choice_input = gr.Dropdown(choices=kitchen_choices, label="Select Kitchen for Unique Items", value="Main Kitchen")
        unique_button = gr.Button("Find Unique Items")
        unique_message = gr.Textbox(label="Unique Items Status", interactive=False)
        unique_button.click(find_unique_items, inputs=unique_kitchen_choice_input, outputs=unique_message)

# Launch the Gradio app
app.launch()
```

```python
import gradio as gr

# Sample data for Main Makeup Inventory and Secondary Jewelry Inventory
main_inventory = [
    ("Lipstick", 10, "Lip Products"),
    ("Foundation", 5, "Face Products"),
    ("Eyeliner", 8, "Eye Products"),
    ("Blush", 3, "Face Products"),
    ("Mascara", 6, "Eye Products")
]

secondary_inventory = [
    ("Necklace", 2, "Neck Accessories"),
    ("Earrings", 5, "Ear Accessories"),
    ("Bracelet", 4, "Wrist Accessories"),
    ("Ring", 10, "Finger Accessories"),
    ("Anklet", 3, "Ankle Accessories")
]

# Define the available categories
makeup_categories = ["Lip Products", "Face Products", "Eye Products"]
jewelry_categories = ["Neck Accessories", "Ear Accessories", "Wrist Accessories", "Finger Accessories", "Ankle Accessories"]

# Define the inventory choices
inventory_choices = ["Makeup Inventory", "Jewelry Inventory"]

# Function to add a new product to the selected inventory
def add_product(name, quantity, category, inventory_choice):
    new_product = (name, quantity, category)

    if inventory_choice == "Makeup Inventory":
        main_inventory.append(new_product)
        return f"Product '{name}' added to Makeup Inventory successfully!"
    elif inventory_choice == "Jewelry Inventory":
        secondary_inventory.append(new_product)
        return f"Product '{name}' added to Jewelry Inventory successfully!"

# Function to display all products in the selected inventory
def display_products(inventory_choice):
    if inventory_choice == "Makeup Inventory":
        inventory = main_inventory
    else:
        inventory = secondary_inventory

    if inventory:
        products = "\n".join([f"Product: {item[0]}, Quantity: {item[1]}, Category: {item[2]}" for item in inventory])
        return f"Products in {inventory_choice}:\n{products}"
    else:
        return f"{inventory_choice} is empty!"

# Function to remove duplicates from the selected inventory
def remove_duplicates(inventory_choice):
    if inventory_choice == "Makeup Inventory":
        seen = {}
        global main_inventory
        for item in main_inventory:
            if item[0] not in seen:
                seen[item[0]] = item
        main_inventory = list(seen.values())
        return "Duplicates removed successfully from Makeup Inventory!"
    elif inventory_choice == "Jewelry Inventory":
        seen = {}
        global secondary_inventory
        for item in secondary_inventory:
            if item[0] not in seen:
                seen[item[0]] = item
        secondary_inventory = list(seen.values())
        return "Duplicates removed successfully from Jewelry Inventory!"

# Function to find common products between both inventories
def find_common_products():
    common_products = set(main_inventory).intersection(set(secondary_inventory))
    if common_products:
        products = "\n".join([f"Product: {item[0]}, Quantity: {item[1]}, Category: {item[2]}" for item in common_products])
        return f"Common Products between Makeup and Jewelry Inventory:\n{products}"
    else:
        return "No common products found."

# Function to find unique products based on the selected inventory
def find_unique_products(inventory_choice):
    if inventory_choice == "Makeup Inventory":
        unique_products = set(main_inventory).difference(set(secondary_inventory))
        if unique_products:
            products = "\n".join([f"Product: {item[0]}, Quantity: {item[1]}, Category: {item[2]}" for item in unique_products])
            return f"Products Unique to Makeup Inventory:\n{products}"
        else:
            return "No unique products in Makeup Inventory."
    else:  # For Jewelry Inventory
        unique_products = set(secondary_inventory).difference(set(main_inventory))
        if unique_products:
            products = "\n".join([f"Product: {item[0]}, Quantity: {item[1]}, Category: {item[2]}" for item in unique_products])
            return f"Products Unique to Jewelry Inventory:\n{products}"
        else:
            return "No unique products in Jewelry Inventory."

# Gradio Interface
def inventory_interface(name, quantity, category, inventory_choice):
    add_message = add_product(name, quantity, category, inventory_choice)
    display_message = display_products(inventory_choice)  # Show products from the selected inventory after adding
    return add_message, display_message

def display_inventory_products(inventory_choice):
    return display_products(inventory_choice)

with gr.Blocks() as app:
    gr.Markdown("## Makeup and Jewelry Inventory Management")

    with gr.Row():
        with gr.Column():
            gr.Markdown("### Add New Product")
            name_input = gr.Textbox(label="Product Name")
            quantity_input = gr.Number(label="Quantity", value=1)
            inventory_choice_input = gr.Dropdown(choices=inventory_choices, label="Select Inventory", value="Makeup Inventory")  # Dropdown for inventory choice

            # Display appropriate categories based on the inventory selected
            category_input = gr.Dropdown(label="Category", choices=makeup_categories, value=makeup_categories[0])
            inventory_choice_input.change(lambda x: makeup_categories if x == "Makeup Inventory" else jewelry_categories,
                                           inputs=[inventory_choice_input],
                                           outputs=[category_input])

            add_message = gr.Textbox(label="Add Product Status", interactive=False)
            display_message = gr.Textbox(label="Current Inventory", interactive=False)  # Define this before using

            add_button = gr.Button("Add Product")
            add_button.click(inventory_interface, inputs=[name_input, quantity_input, category_input, inventory_choice_input], outputs=[add_message, display_message])

    with gr.Row():
        gr.Markdown("### Show Products in Inventory")
        display_button = gr.Button("Show Current Products")
        display_button.click(display_inventory_products, inputs=inventory_choice_input, outputs=display_message)

    with gr.Row():
        remove_inventory_choice_input = gr.Dropdown(choices=inventory_choices, label="Select Inventory for Duplicates", value="Makeup Inventory")
        remove_button = gr.Button("Remove Duplicates")
        remove_message = gr.Textbox(label="Remove Duplicates Status", interactive=False)
        remove_button.click(remove_duplicates, inputs=remove_inventory_choice_input, outputs=remove_message)

    with gr.Row():
        common_button = gr.Button("Find Common Products")
        common_message = gr.Textbox(label="Common Products Status", interactive=False)
        common_button.click(find_common_products, outputs=common_message)

    with gr.Row():
        unique_inventory_choice_input = gr.Dropdown(choices=inventory_choices, label="Select Inventory for Unique Products", value="Makeup Inventory")
        unique_button = gr.Button("Find Unique Products")
        unique_message = gr.Textbox(label="Unique Products Status", interactive=False)
        unique_button.click(find_unique_products, inputs=unique_inventory_choice_input, outputs=unique_message)

# Launch the Gradio app
app.launch()
```
